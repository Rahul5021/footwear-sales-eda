**Insight: Brand and category performance is unusually balanced across the dataset**

Grouping sales by brand and category shows no single combination dominating — revenue across the top 10 groups ranges narrowly between ~305,000 and 330,000 USD, and units sold cluster between ~2,500 and 2,700. ASICS (Lifestyle) and New Balance (Training) lead marginally in both revenue and units, but the rankings shift slightly further down the list depending on whether you sort by units sold or revenue — for example, Puma (Lifestyle) and Nike (Training) outrank ASICS (Basketball) on revenue despite having fewer or similar total units, suggesting modestly higher per-unit pricing in those categories.

That said, the overall uniformity across nearly every brand/category pairing is itself the more notable finding — real-world sales data typically shows more pronounced winners and laggards. This suggests the dataset was generated with intentionally balanced distributions across brands and categories, rather than reflecting organic, skewed market demand. As a result, brand/category alone isn't a strong lever for differentiating performance in this dataset — worth checking whether other dimensions (country, sales channel, income level) show clearer patterns.

```python
df['discount_bucket'] = pd.cut(df['discount_percent'], bins = [-1,0,10,20,100], labels = ['No Discount','1-10%', '11-20%', '21%+'])
```

```python
df.groupby(by = 'discount_bucket')[['units_sold','revenue_usd']].mean()
```

**Insight: Discounting does not meaningfully increase purchase volume**

Grouping orders by discount level shows average units sold per order stays nearly constant regardless of discount size (2.49 units at no discount vs. 2.51 units at 21%+ discount — a negligible difference). Meanwhile, average revenue per order declines steadily as discount increases, from ~ $347 at no discount to ~ $245 at 21%+ discount, a drop of roughly 29%. This indicates that within this dataset, discounting functions purely as a revenue reducer rather than a volume driver — customers do not appear to purchase more units in response to larger discounts. Combined with the earlier finding that this dataset shows unusually uniform performance across brand and category, this reinforces that the data likely reflects synthetically generated patterns rather than organic consumer price sensitivity, where discounting would typically show at least some correlation with purchase volume.

**Insight: Customer income level does not predict purchasing behavior**

Testing income level against average price paid, category preference, brand preference, and discount sensitivity all show the same result: no meaningful variation. Average final price ranges only $1.31 across income levels ($120.55–$121.86), category and brand distributions are evenly spread within roughly 5% of each other across all income groups, and discount responsiveness (average units sold per order) shows no consistent pattern by income level, with all values clustering tightly between 2.46 and 2.55 regardless of discount size. In real consumer markets, income level typically predicts at least some purchasing behavior — such as price sensitivity or brand preference. Its complete absence here, combined with the earlier findings on brand/category, discount response, and channel/country splits, provides strong, consistent evidence that this dataset was synthetically generated with uniformly randomized distributions across all customer segments, rather than modeling real consumer behavior.

**Insight: No meaningful trend across time — confirming the dataset is synthetic**

Yearly revenue shows no growth or decline pattern across 2018–2026, fluctuating narrowly between $970K and $1.03M with no discernible trajectory. Notably, 2020 posts the highest revenue in the dataset ($1.03M) rather than a dip, despite this being a year when real-world footwear retail was significantly disrupted by COVID-19 store closures — a strong indicator the dates were randomly assigned rather than modeled on real market conditions. Monthly totals are similarly flat, with no visible holiday-season spike in November/December that real footwear sales typically show. Combined with the earlier findings across brand, category, discount response, geography, sales channel, and income level — all of which showed no meaningful variation — this provides comprehensive evidence that the dataset was generated synthetically with uniformly randomized values across every dimension tested, rather than reflecting real consumer purchasing behavior.
